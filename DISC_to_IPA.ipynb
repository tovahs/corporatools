```python
# Open File
wordlist = []

# Replace this with the path to your file - code is configured to take second row
# of input tab seperated values file -
with open("Word Lists/1_to_1_words.tsv") as f:
    for line in f:
        #note that I'm choosing to strip out '"' since I don't need that information and it causes errors
        wordlist.append(line.strip('\n').replace('"', '').split("\t"))
```

```python
# Mapping between DISC and IPA - add characters as needed
DISC_2_IPA_dict = {'p': 'p', 'b': 'b', 't': 't', 'd': 'd', 'k': 'k', 'g': 'g', 'N': 'ŋ', 'm': 'm', 'n': 'n', 'l': 'l', 'r': 'r', 'f': 'f', 'v': 'v', 'D': 'ð', 's': 's', 'T': 'θ', 'z': 'z', 'S': 'ʃ', 'Z': 'ʒ', 'j': 'j', 'x': 'x', 'h': 'h', 'w': 'w', 'J': 'tʃ', '_': 'dʒ', 'H': 'n̩', 'C': 'ŋ̩', 'F': 'm̩', 'P': 'l̩', 'I': 'ɪ', 'E': 'ɛ', '{': 'æ', 'V': 'ʌ', 'Q': 'ɒ', 'U': 'ʊ', '@': 'ə', 'i': 'iː', '#': 'ɑː', '$': 'ɔː', 'u': 'uː', '3': 'ɜː', '1': 'eɪ', '2': 'aɪ', '4': 'ɔɪ', '5': 'əʊ', '6': 'aʊ', '7': 'ɪə', '8': 'ɛə', '9': 'ʊə', 'c': 'æ', 'q': 'ɑ̃ː', '0': 'æ̃ː', '˜': 'ɒ̃ː', '"': '"', '-': '-', 'R': '*'}
```

```python
# Write new file transcriptions.tsv and split each syllable by tab
with open("transcriptions.tsv", "w+") as file:
    for word in wordlist:
        
        # convert symbols to IPA
        # Modify this to the row that your transcriptions are in
        transcription = ''.join([DISC_IPA_dict[i] for i in word[1]]).split("-")
        # print("\t".join(transcription))
        
        # write file
        file.write("\t".join(transcription))
        file.write("\n")
```
