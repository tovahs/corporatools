```python
# Building New CSV - containing info from all different sources
# Columns are: Vocalized Noun, Root, Definition, Gender, Frequency, Pattern, Irregular Feminine

# Create csv
f = open("noun_info.csv", "w")
column_names = "Word, Root, Gender, Number, Broken_Plural, Masculin_Plural, Definition \n"

f.write(column_names)

# 0, 5, 9, 12, 14, 19, 30

# Connect to larger Arrmooz database
arramooz_nouns = arramooz.arabicdictionary.ArabicDictionary('nouns')

# itterate through each item
for word in vocal_noun_only:
    print("on word: " + word)
    # Create string that will be written to csv
    line = ""
    
    # Lookup word in Arramooz
    arrmooz_entry = arramooz_nouns.lookup(word)
    
    if len(arrmooz_entry) > 0:
           # Write out variables (this is optomized for code legibility, not performance - change it once)
        word = arrmooz_entry[0][1]
        print(word)
        root = arrmooz_entry[0][6]
        print(root)
        gender = arrmooz_entry[0][10]
        print(gender)
        number = arrmooz_entry[0][13]
        print(number)
        broken_plural = arrmooz_entry[0][15]
        print(broken_plural)
        masculine_plural = str(arrmooz_entry[0][20])
        print(masculine_plural)
        definition = arrmooz_entry[0][31]
        line = word + " ," + root + " ," + gender + " ," + number + " ," + broken_plural + " ," + masculine_plural + " ," + definition + "\n"
    
    else:
        line = "NA, , , , , ,\n"
 
    
    
    f.write(line)

f.close()
```
